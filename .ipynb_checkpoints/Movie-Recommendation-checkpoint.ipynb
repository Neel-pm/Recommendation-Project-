```python
movie = movie[['genres',
'id',
'keywords',
'title',
'overview',
'cast',
'crew',
'production_companies']]
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
import ast
```

```python
#for removing extra brackets from column
def convert(text):
    L =[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movie['genres'] = movie['genres'].apply(convert)
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head()
```

```python
movie['cast'][0]
```

```python
movie['cast'] = movie['cast'].apply(convert)
movie.head()
```

```python
#only need for top 5 cast of movie
movie['cast'] = movie['cast'].apply(lambda x:x[0:5])
```

```python
movie['cast'][0]
```

```python
movie['crew'][0]
```

```python
#only need for director from crew 
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movie['production_companies'] = movie['production_companies'].apply(convert)
movie.head()
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
movie['production_companies'][100]
```

```python
#removing space from the name so that one name will not define as two or more
def collapse(L):
    l1 =[]
    for i in L:
        l1.append(i.replace(' ',''))
    return l1
```

```python
movie['genres'] = movie['genres'].apply(collapse)
```

```python
movie['keywords'] = movie['keywords'].apply(collapse)
```

```python
movie['cast'] = movie['cast'].apply(collapse)
```

```python
movie['crew'] = movie['crew'].apply(collapse)
```

```python
movie['production_companies'] = movie['production_companies'].apply(collapse)
```

```python
movie.head(1)
```

```python
#seprating the each word of overview for the vectorization
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie['tags'] = movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']+movie['production_companies']
```

```python
movie['tags'] = movie['tags'].apply(lambda x: ' '.join(x))
```

```python
movie.head(1)
```

```python
movie = movie.drop(columns=['overview','genres','keywords','cast','crew','production_companies'])
```

```python
movie.head()
```

```python
movie['tags'][0]
```

```python
#perform vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vector = cv.fit_transform(movie['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out().tolist()
#need to remove similar word meaning
```

```python
vector = cv.fit_transform(movie['tags']).toarray()
```

```python
#new tags without similar words
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[1]
```

```python
#recomandation fuction
```

```python
def reco(movies):
    index = movie[movie['title'] == movies].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x : x[1])
    for i in distances[1:6]:
        print(movie.iloc[i[0]].title)
```

```python
reco('Gandhi')
```
